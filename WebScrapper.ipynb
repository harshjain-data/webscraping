```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL = "https://www.zippia.com/job-search/foreign-worker-jobs/top-companies/"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
results = soup.find("div", {"class": "remoteCompaniesList"})
```

```python
results.prettify()
```

```python
job_elements = results.find_all("div", class_="companyCard_companyCard__oGst2")
```

```python
for job_element in job_elements:
    job_element, end="\n"*2
```

```python
for job_element in job_elements:
    company_element = job_element.find("h2", class_="blueOpenSans")
    print(company_element)
    print()
```

```python
companies = []
for job_element in job_elements:
    company_element = job_element.find("h2", class_="blueOpenSans")
    print(company_element.text.strip())
    companies.append(company_element.text.strip())
    print()
```

```python
mylist = pd.DataFrame(companies)
mylist
```
